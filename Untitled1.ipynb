```python
import tkinter as tk
from tkinter import messagebox


def calculate_budget():
    try:
        income = float(income_entry.get())
        rent = float(rent_entry.get())
        food = float(food_entry.get())
        transportation = float(transportation_entry.get())
        other = float(other_entry.get())

        total_expenses = rent + food + transportation + other
        money_left = income - total_expenses

        result_text = (
            f"Total Expenses: ${total_expenses:.2f}\n"
            f"Money Left Over: ${money_left:.2f}"
        )

        if money_left > 0:
            result_text += "\n\nGreat job! You are staying within your budget."
        elif money_left == 0:
            result_text += "\n\nYou broke even this month."
        else:
            result_text += "\n\nCareful! Your expenses are higher than your income."

        result_label.config(text=result_text)

    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter numbers only.")


def clear_fields():
    income_entry.delete(0, tk.END)
    rent_entry.delete(0, tk.END)
    food_entry.delete(0, tk.END)
    transportation_entry.delete(0, tk.END)
    other_entry.delete(0, tk.END)
    result_label.config(text="")


def show_about():
    messagebox.showinfo(
        "About",
        "Simple Budget Tracker\n\n"
        "Enter your monthly income and expenses to calculate your leftover money."
    )


root = tk.Tk()
root.title("Simple Budget Tracker")
root.geometry("430x500")
root.configure(bg="#fff4b8")

menu_bar = tk.Menu(root)
root.config(menu=menu_bar)

help_menu = tk.Menu(menu_bar, tearoff=0)
menu_bar.add_cascade(label="Help", menu=help_menu)
help_menu.add_command(label="About", command=show_about)

title_label = tk.Label(
    root,
    text="Simple Budget Tracker",
    font=("Arial", 20, "bold"),
    fg="#7a4f00",
    bg="#fff4b8"
)
title_label.pack(pady=15)

instructions_label = tk.Label(
    root,
    text="Enter your monthly income and expenses below.",
    font=("Arial", 11),
    fg="#5c4200",
    bg="#fff4b8"
)
instructions_label.pack(pady=5)


def create_input_label(text):
    label = tk.Label(
        root,
        text=text,
        font=("Arial", 12),
        fg="#5c4200",
        bg="#fff4b8"
    )
    label.pack()


def create_entry_box():
    entry = tk.Entry(
        root,
        font=("Arial", 12),
        width=25,
        bg="#fffbea",
        fg="#3d2c00"
    )
    entry.pack(pady=4)
    return entry


create_input_label("Monthly Income:")
income_entry = create_entry_box()

create_input_label("Rent / Housing:")
rent_entry = create_entry_box()

create_input_label("Food:")
food_entry = create_entry_box()

create_input_label("Transportation:")
transportation_entry = create_entry_box()

create_input_label("Other Expenses:")
other_entry = create_entry_box()

calculate_button = tk.Button(
    root,
    text="Calculate Budget",
    font=("Arial", 12, "bold"),
    bg="#f5b700",
    fg="#3d2c00",
    activebackground="#d99a00",
    activeforeground="white",
    padx=10,
    pady=5,
    command=calculate_budget
)
calculate_button.pack(pady=12)

clear_button = tk.Button(
    root,
    text="Clear",
    font=("Arial", 11),
    bg="#ffe28a",
    fg="#3d2c00",
    activebackground="#ffd35c",
    padx=10,
    pady=4,
    command=clear_fields
)
clear_button.pack()

result_label = tk.Label(
    root,
    text="",
    font=("Arial", 12, "bold"),
    fg="#4d3900",
    bg="#fff4b8",
    justify="center"
)
result_label.pack(pady=20)

root.mainloop()
```
